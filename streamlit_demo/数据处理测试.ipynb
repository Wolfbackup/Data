```python
import pandas as pd
```

```python
df=pd.read_csv('national_supchk.csv')
df = df.dropna(subset=['year','prov_name','prod_type1'],axis=0)
df.reset_index(inplace=True)
df.head()
```

```python
suk_total=len(df)
suk_uqld=len(df[df['is_qld']=='不合格'])
uqld_rate=round(suk_uqld/suk_total,4)*100
uqld_rate
```

```python
options=df["year"].unique(),
options
```

```python
df1=pd.pivot_table(df,index=['prov_name'],values=['id','is_qld'],aggfunc={'id':len,'is_qld':lambda x:(x=='不合格').sum()})
df1.reset_index(inplace=True)
df1['rate']=df1['is_qld']/df1['id']*100
df1.rename(columns={'id':'num','is_qld':'uqld'},inplace=True)
df1
```

```python
#行业不合格率
df1=pd.pivot_table(df,index=['prod_type1'],values=['id','is_qld'],aggfunc={'id':len,'is_qld':lambda x:(x=='不合格').sum()})
df1.reset_index(inplace=True)
df1['rate']=df1['is_qld']/df1['id']*100
df1.rename(columns={'id':'num','is_qld':'uqld'},inplace=True)
df1
```

```python
#产品不合格率
df_prod=pd.pivot_table(df,index=['prod_name1'],values=['id','is_qld'],aggfunc={'id':len,'is_qld':lambda x:(x=='不合格').sum()})
df_prod.reset_index(inplace=True)
df_prod['rate']=df_prod['is_qld']/df_prod['id']*100
df_prod.rename(columns={'prod_name1':'产品名称','id':'抽查次数','is_qld':'不合格数','rate':'不合格发现率'},inplace=True)
df_prod
```
